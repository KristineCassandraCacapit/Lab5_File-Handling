```python
import os

folder = os.path.expanduser("~/Documents/Activity_5_Files")

if not os.path.exists(folder):
    print("Folder not found:", folder)
else:
    files = os.listdir(folder)
    results = []

    for file in files:
        if file.endswith(".txt"):
            path = os.path.join(folder, file)

            with open(path, "r", encoding="utf-8") as f:
                lines = f.readlines()

            clean_lines = [line.strip() for line in lines if line.strip() != ""]

            line_count = len(clean_lines)
            word_count = 0
            char_count = 0

            for line in clean_lines:
                words = line.split()
                word_count += len(words)

                for word in words:
                    char_count += len(word)

            wpl = word_count / line_count if line_count > 0 else 0
            cpw = char_count / word_count if word_count > 0 else 0

            results.append([file, line_count, word_count, char_count, round(wpl, 2), round(cpw, 2)])

    results.sort(key=lambda x: x[4], reverse=True)

    print("\nFilename | Lines | Words | Characters | Words/Line | Chars/Word")
    for r in results:
        print(r)

    if results:
        print("\nMost Dense:", results[0][0])
        print("Least Dense:", results[-1][0])
```

```python
import os
import string

folder = os.path.expanduser("~/Documents/Activity_5_Files")

if not os.path.exists(folder):
    print("Folder not found:", folder)
else:
    files = os.listdir(folder)

    # LOAD STOPWORDS
    stopwords = []
    stop_path = os.path.join(folder, "stopwords.txt")

    if os.path.exists(stop_path):
        with open(stop_path, "r", encoding="utf-8") as f:
            stopwords = [line.strip().lower() for line in f]

    word_freq = {}

    for file in files:
        if file.endswith(".txt"):
            path = os.path.join(folder, file)

            with open(path, "r", encoding="utf-8") as f:
                text = f.read().lower()

            for p in string.punctuation:
                text = text.replace(p, "")

            words = text.split()

            for word in words:
                if word not in stopwords:
                    word_freq[word] = word_freq.get(word, 0) + 1

    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

    print("\nTop Words:")
    for word, count in sorted_words[:10]:
        print(word, ":", count)
```

```python
import os
from datetime import datetime
import shutil

student_id = "2025-0962"
student_name = "Kristine Cassandra T. Cacapit"

base_path = os.path.expanduser("~/Documents/Activity_5_Exercises")
file_to_copy = os.path.join(base_path, f"intro_{student_id}.txt")

backup_dir = os.path.join(base_path, f"backup_{student_id}")

os.makedirs(backup_dir, exist_ok=True)

if not os.path.exists(file_to_copy):
    print("File does not exist:", file_to_copy)
else:
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    backup_name = f"intro_{student_id}_{timestamp}.txt"
    backup_path = os.path.join(backup_dir, backup_name)


    shutil.copy(file_to_copy, backup_path)

    file_size = os.path.getsize(backup_path)

    log_file = os.path.join(backup_dir, f"backup_log_{student_id}.txt")

    with open(log_file, "a", encoding="utf-8") as log:
        log.write(f"Student ID: {student_id}\n")
        log.write(f"Student Name: {student_name}\n")
        log.write(f"Original File: {os.path.basename(file_to_copy)}\n")
        log.write(f"Timestamp: {timestamp}\n")
        log.write(f"File Size: {file_size} bytes\n")
        log.write(f"Backup Path: {backup_path}\n")
        log.write("-" * 40 + "\n")

    print("\nBackup completed for", student_id, "(", student_name, ")")
    print("File saved as:", backup_name)
    print("Log updated at:", backup_dir)
```
